- PyTorch

## Step 3: Load Pre-trained Stable Diffusion Model

A pre-trained Stable Diffusion model is loaded from Hugging Face using the Diffusers library.

The model has already learned patterns from a large collection of image-text data, allowing it to generate images from text prompts without training from scratch.

```python
import torch
from diffusers import StableDiffusionPipeline
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
print("Stable Diffusion model loaded successfully")
```

## Step 4: Generate Image from Text Prompt

The user enters a text prompt describing the required image. Stable Diffusion processes the text input and generates a new AI image using the pre-trained diffusion model.

The generated image is displayed as the output.

```python
prompt = input("Enter your image description: ")
image = pipe(prompt).images[0]
image
```

## Step 5: Save Generated Image

The generated output image is saved as a PNG file so that users can download and use the AI-generated result.

This step stores the final image created by the Stable Diffusion model.

```python
image.save("generated_image.png")
print("Image saved successfully !")
```
